```python
# Importando as bibliotecas necessárias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com as métricas de centralidade para os bairros próximos à UFRN
df = pd.DataFrame({
    'bairro': ['Petrópolis', 'Lagoa Nova', 'Candelária', 'Tirol', 'Alecrim', 'Capim Macio', 'Ponta Negra'],
    'grau_centralidade': [0.85, 0.75, 0.65, 0.9, 0.6, 0.7, 0.55],
    'proximidade_centralidade': [0.9, 0.8, 0.7, 0.85, 0.6, 0.65, 0.5],
    'intermediacao_centralidade': [0.7, 0.65, 0.75, 0.8, 0.6, 0.7, 0.55],
    'eigenvector_centrality': [0.88, 0.82, 0.77, 0.9, 0.65, 0.72, 0.58]
})

# Criação do PairGrid para explorar as métricas de centralidade dos bairros
fig = sns.PairGrid(df.drop(columns='bairro'))  # Remover a coluna 'bairro' para análise das métricas numéricas

# Usar scatterplot nas áreas acima da diagonal para ver a correlação entre as métricas
fig.map_upper(sns.scatterplot)

# Usar KDE (densidade) nas áreas abaixo da diagonal para mostrar as distribuições das métricas
fig.map_lower(sns.kdeplot, cmap="Reds_r")

# Usar KDE nas diagonais para mostrar as distribuições univariadas das métricas
fig.map_diag(sns.kdeplot, lw=2, legend=False)

# Exibir o gráfico
plt.suptitle('Análise Multivariada das Métricas de Centralidade nos Bairros Próximos à UFRN', size=16)
plt.tight_layout()
plt.subplots_adjust(top=0.95)  # Ajustar o título para não sobrepor os gráficos
plt.show()
```
