```python
# Importando as bibliotecas necessrias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Criando o DataFrame com as mtricas de centralidade para os bairros prximos  UFRN
df = pd.DataFrame({
    'bairro': ['Petrpolis', 'Lagoa Nova', 'Candelria', 'Tirol', 'Alecrim', 'Capim Macio', 'Ponta Negra'],
    'grau_centralidade': [0.85, 0.75, 0.65, 0.9, 0.6, 0.7, 0.55],
    'proximidade_centralidade': [0.9, 0.8, 0.7, 0.85, 0.6, 0.65, 0.5],
    'intermediacao_centralidade': [0.7, 0.65, 0.75, 0.8, 0.6, 0.7, 0.55],
    'eigenvector_centrality': [0.88, 0.82, 0.77, 0.9, 0.65, 0.72, 0.58]
})

# Criao do PairGrid para explorar as mtricas de centralidade dos bairros
fig = sns.PairGrid(df.drop(columns='bairro'))  # Remover a coluna 'bairro' para anlise das mtricas numricas

# Usar scatterplot nas reas acima da diagonal para ver a correlao entre as mtricas
fig.map_upper(sns.scatterplot)

# Usar KDE (densidade) nas reas abaixo da diagonal para mostrar as distribuies das mtricas
fig.map_lower(sns.kdeplot, cmap="Reds_r")

# Usar KDE nas diagonais para mostrar as distribuies univariadas das mtricas
fig.map_diag(sns.kdeplot, lw=2, legend=False)

# Exibir o grfico
plt.suptitle('Anlise Multivariada das Mtricas de Centralidade nos Bairros Prximos  UFRN', size=16)
plt.tight_layout()
plt.subplots_adjust(top=0.95)  # Ajustar o ttulo para no sobrepor os grficos
plt.show()
```
