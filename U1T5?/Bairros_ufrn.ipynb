```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Criando uma rede exemplo com conexes entre bairros
G = nx.Graph()

# Adicionando bairros (ns) e conexes principais (arestas)
bairros = ["UFRN", "Lagoa Nova", "Candelria", "Capim Macio", "Ponta Negra", "Nova Descoberta", "Tirol"]
conexoes = [("UFRN", "Lagoa Nova"), ("UFRN", "Candelria"), ("UFRN", "Capim Macio"),
            ("Lagoa Nova", "Nova Descoberta"), ("Candelria", "Capim Macio"),
            ("Capim Macio", "Ponta Negra"), ("Candelria", "Tirol")]

G.add_nodes_from(bairros)
G.add_edges_from(conexoes)

# Degree Centrality
degree_centrality = nx.degree_centrality(G)
nx.set_node_attributes(G, degree_centrality, 'degree_centrality')

# Visualizando a rede destacando Degree Centrality
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G)
nodes = nx.draw_networkx_nodes(G, pos, node_size=[v * 1000 for v in degree_centrality.values()],
                               node_color=list(degree_centrality.values()), cmap='coolwarm')
edges = nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos, font_size=12)
plt.colorbar(nodes)
plt.title("Degree Centrality nos Bairros")
plt.show()

# PDF e CDF para Degree Centrality
values = list(degree_centrality.values())
sns.histplot(values, kde=True, color="blue", stat="density", label="PDF")
sns.ecdfplot(values, color="orange", label="CDF")
plt.legend()
plt.title("PDF e CDF da Degree Centrality")
plt.xlabel("Degree Centrality")
plt.ylabel("Density")
plt.show()
```
