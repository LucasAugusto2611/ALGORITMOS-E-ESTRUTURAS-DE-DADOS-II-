```python
import networkx as nx
import matplotlib.pyplot as plt

# Definindo os bairros prximos  UFRN
bairros = ['Petrpolis', 'Lagoa Nova', 'Candelria', 'Tirol', 'Alecrim', 'Capim Macio', 'Ponta Negra']

# Criando um grafo (rede) com os bairros como ns
G = nx.Graph()

# Adicionando os bairros como ns
G.add_nodes_from(bairros)

# Definindo as conexes (arestas) entre os bairros
arestas = [
    ('Petrpolis', 'Lagoa Nova'),
    ('Petrpolis', 'Tirol'),
    ('Lagoa Nova', 'Candelria'),
    ('Lagoa Nova', 'Petrpolis'),
    ('Tirol', 'Alecrim'),
    ('Tirol', 'Petrpolis'),
    ('Alecrim', 'Capim Macio'),
    ('Capim Macio', 'Ponta Negra'),
    ('Capim Macio', 'Alecrim')
]

# Adicionando as arestas ao grafo
G.add_edges_from(arestas)

# Visualizando a rede dos bairros
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold')
plt.title("Rede de Bairros Prximos  UFRN")
plt.show()

# Calculando o nmero de cores (k-core) de cada n
core_numbers = nx.core_number(G)

# Exibindo os nmeros de core de cada bairro
print("Core numbers:", core_numbers)

# Identificando os valores distintos de k-core (os diferentes k-cores na rede)
k_cores = set(core_numbers.values())
print("Valores distintos de k-core:", k_cores)

# Quantidade de k-cores diferentes
print(f"Quantidade de k-cores diferentes: {len(k_cores)}")

# Para um k-core especfico, por exemplo, k=3:
k_core_3 = nx.k_core(G, k=3)
print("Ns no 3-core:", k_core_3.nodes())


```
