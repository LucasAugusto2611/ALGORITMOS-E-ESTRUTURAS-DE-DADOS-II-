```python
import networkx as nx
import matplotlib.pyplot as plt

# Definindo os bairros próximos à UFRN
bairros = ['Petrópolis', 'Lagoa Nova', 'Candelária', 'Tirol', 'Alecrim', 'Capim Macio', 'Ponta Negra']

# Criando um grafo (rede) com os bairros como nós
G = nx.Graph()

# Adicionando os bairros como nós
G.add_nodes_from(bairros)

# Definindo as conexões (arestas) entre os bairros
arestas = [
    ('Petrópolis', 'Lagoa Nova'),
    ('Petrópolis', 'Tirol'),
    ('Lagoa Nova', 'Candelária'),
    ('Lagoa Nova', 'Petrópolis'),
    ('Tirol', 'Alecrim'),
    ('Tirol', 'Petrópolis'),
    ('Alecrim', 'Capim Macio'),
    ('Capim Macio', 'Ponta Negra'),
    ('Capim Macio', 'Alecrim')
]

# Adicionando as arestas ao grafo
G.add_edges_from(arestas)

# Visualizando a rede dos bairros
plt.figure(figsize=(8, 6))
nx.draw(G, with_labels=True, node_size=3000, node_color='lightblue', font_size=12, font_weight='bold')
plt.title("Rede de Bairros Próximos à UFRN")
plt.show()

# Calculando o número de cores (k-core) de cada nó
core_numbers = nx.core_number(G)

# Exibindo os números de core de cada bairro
print("Core numbers:", core_numbers)

# Identificando os valores distintos de k-core (os diferentes k-cores na rede)
k_cores = set(core_numbers.values())
print("Valores distintos de k-core:", k_cores)

# Quantidade de k-cores diferentes
print(f"Quantidade de k-cores diferentes: {len(k_cores)}")

# Para um k-core específico, por exemplo, k=3:
k_core_3 = nx.k_core(G, k=3)
print("Nós no 3-core:", k_core_3.nodes())


```
