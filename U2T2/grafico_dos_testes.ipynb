```python
import numpy as np
import matplotlib.pyplot as plt

# Dados de tempo de execuo
sizes = [10, 100, 1000, 10000, 100000]
times_closest = [0.000002, 0.000005, 0.000033, 0.000326, 0.003098]  # Tempo de execuo para solver_closest
times_kth_largest = [0.000006, 0.000018, 0.000186, 0.002428, 0.030387]  # Tempo de execuo para solver_kth_largest

# Plotando o grfico
plt.figure(figsize=(10, 6))

# Grfico para o solver_closest
plt.plot(sizes, times_closest, '-o', label='Solver Closest')

# Grfico para o solver_kth_largest
plt.plot(sizes, times_kth_largest, '-o', label='Solver Kth Largest')

# Ajustando o grfico
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Tamanho do Vetor (log)')
plt.ylabel('Tempo de Execuo (log)')
plt.title('Desempenho dos Algoritmos')
plt.legend()
plt.grid(True)

# Exibindo o grfico
plt.show()

```
