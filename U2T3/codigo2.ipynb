```python
import networkx as nx
import matplotlib.pyplot as plt

# Criação do grafo com pontos de interesse (PoI) e custos entre eles
def criar_grafo():
    grafo = nx.Graph()
    pontos = [
        'Aeroporto',
        'Rodoviária',
        'Praia do Forte',
        'Ponta Negra',
        'Centro Histórico'
    ]

    # Adicionando arestas com pesos (distância ou custo estimado)
    arestas = [
        ('Aeroporto', 'Rodoviária', 12),
        ('Aeroporto', 'Praia do Forte', 18),
        ('Rodoviária', 'Centro Histórico', 5),
        ('Centro Histórico', 'Praia do Forte', 10),
        ('Ponta Negra', 'Centro Histórico', 8),
        ('Ponta Negra', 'Rodoviária', 14),
        ('Praia do Forte', 'Ponta Negra', 20)
    ]

    grafo.add_weighted_edges_from(arestas)
    return grafo

# Função para calcular a MST usando o algoritmo de Kruskal
def calcular_mst(grafo):
    mst = nx.minimum_spanning_tree(grafo, weight='weight', algorithm='kruskal')
    return mst

# Função para visualizar o grafo e a MST
def visualizar_grafos(grafo, mst):
    pos = nx.spring_layout(grafo)  # Posicionamento dos nós para visualização (você pode testar outros layouts)

    # Grafo original
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=12)
    nx.draw_networkx_edge_labels(grafo, pos, edge_labels=nx.get_edge_attributes(grafo, 'weight'))
    plt.title("Grafo Original")

    # Árvore Geradora Mínima (MST)
    plt.subplot(1, 2, 2)
    nx.draw(mst, pos, with_labels=True, node_color='lightgreen', edge_color='red', width=2, node_size=2000, font_size=12)
    nx.draw_networkx_edge_labels(mst, pos, edge_labels=nx.get_edge_attributes(mst, 'weight'))
    plt.title("Árvore Geradora Mínima (MST)")

    plt.tight_layout()
    plt.show()

# Função principal para executar o trabalho
def main():
    print("Criando o grafo...")
    grafo = criar_grafo()

    print("\nCalculando a Árvore Geradora Mínima (MST)...")
    mst = calcular_mst(grafo)

    print("\nVisualizando os grafos...")
    visualizar_grafos(grafo, mst)

    # Exibindo os detalhes da MST
    total_peso = sum(nx.get_edge_attributes(mst, 'weight').values())
    print(f"Peso total da MST: {total_peso} unidades de custo/distance")
    print("Arestas na MST:")
    for aresta in mst.edges(data=True):
        print(f"{aresta[0]} - {aresta[1]}: {aresta[2]['weight']}")

if __name__ == "__main__":
    main()
```
